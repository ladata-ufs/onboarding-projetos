### Importações e Reexportações de café por país

```python
# Média
importacao['media'] = importacao['Total_import'] / 30

# Barplot
plt.figure(figsize=(20, 6))
sns.barplot(x='Country', y='media', data=importacao)
plt.xticks(rotation=65)
plt.title('Média total de Importação por país')
plt.show()
```

```python
# Média
re_exportacao['media'] = re_exportacao['Total_re_export'] / 30

# Barplot
plt.figure(figsize=(20, 6))
sns.barplot(x='Country', y='media', data=re_exportacao)
plt.xticks(rotation=65)
plt.title('Média total de Reexportação por país')
plt.show()
```

Observando os gráficos acima, é possível notar que os Estados Unidos são o maior importador de café, enquanto a Alemanha é o maior reexportador.

Países como Japão e Itália são grandes importadores, mas não aparecem tão alto na lista de reexportadores. Isso sugere que esses países consomem a maior parte do café importado internamente, com menor reexportação.


### Top 20 maiores importadores e reexportadores de café

```python
# Ordenar de forma decrescente
df_sorted = importacao.sort_values('Total_import', ascending=False)

# Selecionando os 20 maiores países sem duplicatas
top_20_importadores = df_sorted['Country'].drop_duplicates().head(20)

# Mostrar os 20 maiores produtores
print("Top 20 maiores importadores de café")
print(top_20_importadores)
```

```python
# Ordenar de forma decrescente
df_sorted = re_exportacao.sort_values('Total_re_export', ascending=False)

# Selecionando os 20 maiores países sem duplicatas
top_20_reexportadores = df_sorted['Country'].drop_duplicates().head(20)

# Mostrar os 20 maiores reexportadores
print("Top 20 maiores reexportadores de café")
print(top_20_reexportadores)
```

### Conclusão
Baseado nas informações acima é possível concluir que países como Alemanha, Bélgica e Estados Unidos não apenas importam grandes quantidades de café para consumo interno, mas também desempenham um papel crucial na redistribuição para outros mercados. Isso sugere que esses países possuem uma infraestrutura logística e comercial bem desenvolvida, capaz de gerenciar grandes volumes de comércio internacional. Por outro lado, países como Japão e Itália são principalmente consumidores, com uma menor participação na reexportação, refletindo um alto consumo interno.
