Inicialemnte para que fosse possível a análise e manipulação dos dados presentes no DataSet em questão, foi necessário o uso de uma biblioteca bastante conhecida, a biblioteca PANDAS, que é focada em análise de dados, sendo peça chave para a conclusão do projeto como todo. Além do PANDAS, outra biblioteca utilizada foi a Matplotlib, para tornar a visualização dos resultados mais Objetiva e Agradável.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Com as bibliotecas já importadas, o próximo passo é definir o Data Frame de nosso interesse. (Em casos do arquivo do Data Set não estar na mesma pasta do Sketch, basta utilizar o caminho do diretório onde o arquivo se encontra.)

```python
df_poke_lend = pd.read_csv("pokemon.csv")
```

A coluna que informa se o Pokémon é Lendário ou não, é Binário, ou seja, NÃO_LENDÁRIO = 0, LENDÁRIO = 1, então, para saber quantos pokémons lendários a geração em questão possui, há duas formas, usando uma função de contagem, ou de somatório da coluna 'is_legendary', opção utilizada nesse caso.  

```python
soma_lendarios_por_geracao = df_poke_lend.groupby('generation')['is_legendary'].sum()
```

Fazemos então um laço para que esse somatório se aplique a cada geração e retornanmos um print com o quantitivo de Pokémons Lendários em cada um delas. (Nesse caso, como terá a plotagem do gráfico, printar a quantidade serve mais como um complemento a visualização do resultado) 

```python
for geracao, soma_lendarios in soma_lendarios_por_geracao.items():
    print("A {}ª geração de Pokémon possui {} pokémons \33[33mLENDÁRIOS\33[0m".format(geracao, soma_lendarios))
```

Por fim, foi realizada a plotagem do gráfico, sendo que x = Geração do Pokémon e y = Somatório de Pokémon Lendários em cada Geração 

```python
soma_lendarios_por_geracao.plot.bar(x = geracao, y = soma_lendarios)
plt.title('Quantidade de Pokémons Lendários em cada Geração')
plt.xlabel("Gerações")
plt.ylabel("Pokémons Lendários")
plt.xticks(rotation = 0)
plt.show()
```
